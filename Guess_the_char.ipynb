```python
import random

def play_game():

    print("Welcome to the Word Guessing Game!")
    print("You will need to guess a word by finding its letters within a given name.\n")
    print("********* Let's get started! *********")
    print("Explanation of the Game")
    print("A word will be randomly selected and hidden within a given name.")
    print("You have 20 chances to guess the word.")
    print("For each guess, I will tell you how many letters you guessed correctly.")
    print("Try to guess the word within the given chances!\n")

    name = input("Enter your name: ")
    print(f"\nHello, {name}!\n")

    words = ["apple", "banana", "cherry", "watermelon", "elderberry", "avocado", "grape", "orange", "jackfruit", "kiwi"]
    word = random.choice(words).lower()

    chances = 20
    guessed_letters = []
    word_found = False

    while chances > 0 and not word_found:
        print(f"You have chances left: {chances}\n")

        guess = input("Enter your guess: ").lower()
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter. Try again!")
            continue

        guessed_letters.append(guess)

        if guess in word:
            print("Correct guess!")
        else:
            print("Incorrect guess!")

        word_progress = ""
        for letter in word:
            if letter in guessed_letters:
                word_progress += letter
            else:
                word_progress += "_"

        print(f"Word progress: {word_progress}\n")

        if word_progress == word:
            word_found = True
            print("Congratulations! You guessed the word correctly!")
        else:
            chances -= 1

    if not word_found:
        print("Sorry, you ran out of chances.")
        print(f"The word was: {word}")

    play_again = input("Would you like to play again? (yes/no): ")
    if play_again.lower() == "yes":
        print("\nLet's play again!\n")
        play_game()
    else:
        print("Thank you for playing the Word Guessing Game!")

play_game()
```
